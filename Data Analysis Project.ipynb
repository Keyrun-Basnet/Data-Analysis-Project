# Importing Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

# Write a python program to merge data from each month into one CSV and read in updated dataframe.

```python
CSV_folder = 'C:\\Users\\kiran\\OneDrive\\Documents\\Smart Data Discovery\\Coursework\\csv'
CSV_files = os.listdir(CSV_folder)

merged_CSV = []
for each in CSV_files:
    A = pd.read_csv('C:/Users/kiran/OneDrive/Documents/Smart Data Discovery/Coursework/csv/' + each)
    merged_CSV.append(A)
merged_CSV = pd.concat(merged_CSV)
merged_CSV
#merged_CSV.to_csv("merged_CSV.csv", index = False)     this line of code is for creating a CSV file of the merged CSVs.
```

# Write a python program to remove the NaN missing values from updated dataframe.

```python
merged_CSV = merged_CSV.dropna()
merged_CSV
```

# Write a python program to convert Quantity Ordered and Price Each to numeric.

```python
pd.to_numeric(pd.Series(merged_CSV['Quantity Ordered']))
pd.to_numeric(pd.Series(merged_CSV['Price Each']))
merged_CSV
```

# Create a new column named Month from Ordered Date of updated dataframe and convert it to integer as data type.

```python
merged_CSV['Month'] = pd.to_datetime(merged_CSV['Order Date']).dt.month
merged_CSV['Month'].astype(int)
merged_CSV
```

```python
merged_CSV.dtypes
```

# Create a new column named City from Purchase Address based on the value in updated dataframe.

```python
address = merged_CSV['Purchase Address']
citylist = []
for each in address:
    city = each.split(",")[1].strip()
    citylist.append(city)
merged_CSV['City'] = citylist
merged_CSV
```

# Which Month has the best sales? and how much was the earning in that month? Make a bar graph of sales as well.

```python
merged_CSV['Total Sales'] = merged_CSV['Quantity Ordered']*merged_CSV['Price Each']
merged_CSV
```

```python
monthly_Sales = merged_CSV.groupby(merged_CSV['Month'])['Total Sales'].sum()
monthly_Sales
```

```python
monthly_Sales.sort_values(ascending = False).head(1)
```

```python
monthly_Sales.plot.bar()
plt.title('Total Sales')
plt.ylabel('Total Sales in Millions')
plt.show()
```

# Which city has sold the highest product?

```python
city_Sales = merged_CSV.groupby(merged_CSV['City'])['Quantity Ordered'].sum()
city_Sales
```

```python
city_Sales.sort_values(ascending = False).head(1)
```

# Which product was sold the most in overall? Illustrate it through bar graph.

```python
Product_Sold = merged_CSV.groupby(merged_CSV['Product'])['Quantity Ordered'].sum().sort_values(ascending = False)
Product_Sold
```

```python
Product_Sold.plot.bar()
plt.title('Quantity Ordered')
plt.ylabel('Total Sales in Millions')
plt.show()
```
